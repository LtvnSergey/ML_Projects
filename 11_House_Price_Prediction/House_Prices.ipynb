# Overview

Ask a home buyer to describe their dream house, and they probably won't begin with the height of the basement ceiling or the proximity to an east-west railroad. But this playground competition's dataset proves that much more influences price negotiations than the number of bedrooms or a white-picket fence.

With 79 explanatory variables describing (almost) every aspect of residential homes in Ames, Iowa, this competition challenges you to predict the final price of each home.

## Functions

```python
def median_rating(input_df, column):
    df = input_df.copy()
    
    median_price = df[['SalePrice', column]].groupby(column).median()
    median_price_dict = median_price['SalePrice'].to_dict()
    df[column+'_median_price'] = df[column].map(median_price_dict)
    
    rating = dict(sorted(median_price_dict.items(), 
                         key=lambda item:item[1]))
    
    i=0
    for key in rating.keys():
        rating[key] = i
        i+=1
        
    return rating


def box_plots_rating(input_df, column, rating):

    df = input_df.copy()    
    df[column+' median price'] = df[column].map(rating)
    
    fig, ax = plt.subplots(figsize=(15, 20))

    ax = sns.boxplot(x='SalePrice', 
                     y=column, 
                     data=df.sort_values(by=column+' median price', ascending=False)
                    )
    ax.grid()
    ax.set_title(column+' rating based on median sale price')
    plt.show()
    
```

```python
def train_models(estimator, X_train, y_train, X_vaild, y_valid):
    
    estimator.fit(X_train, y_train)
    y_pred = estimator.predict(X_vaild)
    metric_r2 = round(r2_score(y_valid, y_pred), 2)
    
    metric_rmse = round(mean_squared_error(y_valid, y_pred, squared=False), 2)
    
    results = pd.DataFrame(data=[[estimator.__class__.__name__, metric_r2, metric_rmse]], 
                           columns=['Estimator', 'R2', 'RMSE']).set_index(keys='Estimator')
    display(results)
    return estimator
```

#### Categories encoding

##### Encoding by quality

- For some categorical features we can encode categories to score when its clearly that the better the condition the higher the sale price and vice versa  


- For example: 

    column 'ExterCond' - Evaluates the present condition of the material on the exterior

    values:
       Ex	Excellent         - score 4
       Gd	Good              - score 3
       TA	Average/Typical   - score 2
       Fa	Fair              - score 1
       Po	Poor              - score 0

```python
quality_columns = [
                   'ExterQual',
                   'ExterCond',
                   'BsmtQual',
                   'BsmtCond',
                   'BsmtFinType1',
                   'BsmtFinType2',
                   'HeatingQC',
                   'KitchenQual',
                   'Functional',
                   'FireplaceQu',
                   'GarageFinish',
                   'GarageQual',
                   'GarageCond',
                   'PoolQC',
                   'Fence'
                  ]

remain_quality_col = []

print('Quality columns for encoding:')

for col in quality_columns:
    print('\n'+col+':')
    try:
        print(list(train_df[train_df[col].isna()==False][col].unique()))
        remain_quality_col.append(col)
    except:
        print('-droped-')
```

```python
dict_quality = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}
dict_type = {'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0}
dict_finish = {'Fin':3, 'RFn':2, 'Unf':1, 'NA':0} 

dict_quality_encoding = {**dict_quality, **dict_type, **dict_finish}
```

```python
train_df[remain_quality_col]
```

```python
for col in remain_quality_col:
    train_df[col] = train_df[col].map(dict_quality_encoding, na_action='ignore')

train_df[remain_quality_col]
```
